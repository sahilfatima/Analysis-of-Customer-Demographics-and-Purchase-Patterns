```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading Dataset
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
#Viewing summary of Null values in Data
df.dropna().isnull().sum()
```

```python
# Dropping rows with missing 'Gender' or 'Purchased' values
data = df.dropna(subset=['Gender', 'Purchased'])

# Counting the number of male and female purchasers
gender_counts = data[df['Purchased'] == 1]['Gender'].value_counts()

# Number of Purchases by Gender
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Number of Purchases')
plt.title('Number of Purchases by Gender')
plt.show()
```

```python
# Distribution of Ages -> Plotting histogram
data['Age'].plot(kind='hist', bins=10, color='green')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

```python
#Age vs. Estimated Salary
plt.scatter(data['Age'], data['EstimatedSalary'], color='purple', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Age vs. Estimated Salary')
plt.show()
```

```python
# Estimated Salary by Gender
data.boxplot(column='EstimatedSalary', by='Gender', grid=False)
plt.xlabel('Gender')
plt.ylabel('Estimated Salary')
plt.title('Estimated Salary by Gender')
plt.suptitle('')  # Suppress the default title
plt.show()
```

```python
# Counting the number of purchases and non-purchases
purchase_counts = data['Purchased'].value_counts()

# Proportion of Purchases
purchase_counts.plot(kind='pie', autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Proportion of Purchases')
plt.ylabel('')
plt.show()
```

```python
# Estimated Salary Distribution
data['EstimatedSalary'].plot(kind='hist', bins=10, color='orange')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Estimated Salary')
plt.show()
```

```python
# Create age groups
bins = [18, 25, 35, 45, 55, 65, 75]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Count the number of purchases by age group
age_group_counts = df[df['Purchased'] == 1]['AgeGroup'].value_counts().sort_index()

# Number of Purshases by Age Group
age_group_counts.plot(kind='bar', color='cyan')
plt.xlabel('Age Group')
plt.ylabel('Number of Purchases')
plt.title('Number of Purchases by Age Group')
plt.show()
```
